## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_filtered = application_df.drop(columns=['EIN', 'NAME'])
application_df_filtered.head()
```

```python
# Determine the number of unique values in each column
unique_counts = application_df_filtered.nunique()

# Print the unique counts
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type = application_df_filtered['APPLICATION_TYPE'].value_counts()
application_type
```

```python
# Determine the value counts of APPLICATION_TYPE
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Choose a cutoff value  for the number of occurrences
cutoff_value = 500

# Create a list of application types to be replaced with "Other"
application_types_to_replace = list(application_type_counts[application_type_counts < cutoff_value].index)

# Replace in the dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df_filtered['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# Filter for CLASSIFICATION value counts > 1
classification_counts_greater_than_1 = classification_counts[classification_counts > 1]
print(classification_counts_greater_than_1)
```

```python
# Determine the value counts of CLASSIFICATION
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Choose a cutoff value for the number of occurrences
cutoff_value = 1000

# Create a list of classifications to be replaced with "Other"
classifications_to_replace = list(classification_counts[classification_counts < cutoff_value].index)

# Replace in the dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# List of categorical columns
categorical_columns = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']

# Apply pd.get_dummies to convert categorical variables to dummy/indicator variables
application_df_encoded = pd.get_dummies(application_df_filtered, columns=categorical_columns)

# Display the first few rows of the encoded DataFrame
application_df_encoded.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Define features (X) and target (y)
X = application_df_encoded.drop(columns=['IS_SUCCESSFUL'])
y = application_df_encoded['IS_SUCCESSFUL']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# Define the number of input features
n_features = X_train.shape[1]



# Define the number of nodes in the hidden layers
hidden_layer1_nodes = 80
hidden_layer2_nodes = 30

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer1_nodes, input_dim=n_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer2_nodes, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()


```

```python
# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
nn.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
# Train the model
nn.fit(X_train_scaled, y_train, epochs=100, verbose=2)
```

## Compile, Train and Evaluate the Model

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('model.h5')
```
